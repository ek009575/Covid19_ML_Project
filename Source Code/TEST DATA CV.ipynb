```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

# Load dataset
uk_updated = pd.read_csv("last.csv")
uk_updated['positivity_rate'] = uk_updated['daily_cases'] / (uk_updated['new_pcr_test'])
uk_updated['daily_cases_lag7'] = uk_updated['daily_cases'].shift(7)
uk_updated['new_pcr_test_7day_avg'] = uk_updated['new_pcr_test'].rolling(window=7).mean()
uk_updated['daily_cases_lag14'] = uk_updated['daily_cases'].shift(14)
uk_updated = uk_updated.fillna(0)
uk_updated['date'] = pd.to_datetime(uk_updated['date'])
uk_updated['ds'] = uk_updated['date']
uk_updated['y'] = uk_updated['cum_cases']
uk_updated['days'] = (uk_updated['date'] - uk_updated['date'].min()).dt.days

# Define the logistic model function
def logistic_model(t, a, b, K):
    return K / (1 + np.exp(-b * (t - a)))

# Calculating Q_top
def calculate_q_top(data):
    params, _ = curve_fit(logistic_model, data['days'], data['y'], maxfev=10000)
    a, b, c = params
    t_fast = int(a)
    current_day = data['days'].max()
    
    if current_day > t_fast:
        t_max = current_day + 30
    else:
        t_max = t_fast * 1.5
    
    Q_top = logistic_model(t_max, *params)
    return Q_top

# divide datasets for prophet 
train_start_date = '2020-01-03'
train_end_date = '2023-05-31'
validation_start_date = '2022-07-01'
validation_end_date = '2023-05-31'
test_end_date = '2023-09-02'

# History column convert date format
uk_updated['date'] = pd.to_datetime(uk_updated['date'])

# create masks for datasets
train_data = uk_updated[(uk_updated['date'] >= train_start_date) & (uk_updated['date'] <= train_end_date)]
val_data = uk_updated[(uk_updated['date'] >= train_start_date) & (uk_updated['date'] <= validation_end_date)]
test_data = uk_updated[(uk_updated['date'] >= validation_start_date) & (uk_updated['date'] <= test_end_date)]

# Q_top calculation
Q_top_train = calculate_q_top(train_data)
Q_top_test = calculate_q_top(test_data) if not test_data.empty else Q_top_train
Q_top_val_data = calculate_q_top(val_data) if not val_data.empty else Q_top_train

def optimize_coverage(horizon_values, period_values):   #the function applying cross validation
    best_coverage = 0
    best_horizon = None
    best_period = None
    best_metrics = None
    
 
    regressors = [
        'cum_pcr_test',
        'new_pcr_test_7day_avg',
        'cum_first_dose',
        'cum_second_dose',
        'cum_third_dose',
        'First_lockdown',
        'Second_lockdown',
        'Third_lockdown',
        'Omicron_Flag',
    ]
    
    initial = 909  # Initial days for 2 yars

    for period in period_values:
        for horizon in horizon_values:
            try:
                # create the model again for each iteration
                model = Prophet(
                    growth='logistic',
                    weekly_seasonality=True,
                    yearly_seasonality=True,
                 
                    interval_width=0.95
                )
                
                for element in regressors:
                    model.add_regressor(element)
                
                model.add_country_holidays(country_name='UK')
                
                # update train and test dataset after iteration
                train_data['cap'] = calculate_q_top(train_data)
                test_data['cap'] = calculate_q_top(test_data)
                
                model.fit(train_data[['date', 'y', 'cap'] + regressors].rename(columns={'date': 'ds'}))
                
                df_cv = cross_validation(model, initial=f'{initial} days', period=f'{period} days', horizon=f'{horizon} days')
                
                df_p = performance_metrics(df_cv)
                
                coverage = df_p['coverage'].mean()
                print(f'Initial: {initial}, Period: {period}, Horizon: {horizon}, Coverage: {coverage}')
                cv_results_df = df_cv[['ds', 'yhat', 'yhat_lower', 'yhat_upper',"y"]]
                print(cv_results_df)
                uk_updated
                if coverage > best_coverage:
                    best_coverage = coverage
                    best_horizon = horizon
                    best_period = period
                    best_metrics = df_p
                    best_results_df=cv_results_df
            except Exception as e:
                print(f'Error with Period: {period}, Horizon: {horizon} - {str(e)}')
    
    return best_period, best_horizon, best_coverage, best_metrics, best_results_df

horizon_values = range(20, 60, 10)  #Day iterations
period_values = range(30, 150, 30)

best_period, best_horizon, best_coverage, best_metrics,best_results_df = optimize_coverage(horizon_values, period_values)

print(f'Best Period: {best_period}, Best Horizon: {best_horizon}, Best Coverage: {best_coverage}')
if best_metrics is not None:   #show performance metrics
    print(best_metrics)

    fig, axes = plt.subplots(4, 1, figsize=(10, 24))

    axes[0].plot(best_metrics['horizon'], best_metrics['mape'], label='MAPE')
    axes[0].set_xlabel('Horizon')
    axes[0].set_ylabel('MAPE')
    axes[0].set_title('Farklı Horizonlarda MAPE')
    axes[0].legend()

    axes[1].plot(best_metrics['horizon'], best_metrics['mae'], label='MAE', color='orange')
    axes[1].set_xlabel('Horizon')
    axes[1].set_ylabel('MAE')
    axes[1].set_title('Farklı Horizonlarda MAE')
    axes[1].legend()

    axes[2].plot(best_metrics['horizon'], best_metrics['rmse'], label='RMSE', color='green')
    axes[2].set_xlabel('Horizon')
    axes[2].set_ylabel('RMSE')
    axes[2].set_title('Farklı Horizonlarda RMSE')
    axes[2].legend()

    axes[3].plot(best_metrics['horizon'], best_metrics['coverage'], label='Coverage', color='red')
    axes[3].set_xlabel('Horizon')
    axes[3].set_ylabel('Coverage')
    axes[3].set_title('Farklı Horizonlarda Coverage')
    axes[3].legend()

    plt.tight_layout()
    plt.show()
else:
    print("No valid metrics found. Please adjust your model or parameters.")
```
